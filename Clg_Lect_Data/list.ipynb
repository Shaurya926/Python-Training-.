```python
# Question: WRP in python to divide the list into 2 equal halves and print the sum of each half.    
lst=[2,4,6,8,3,5,8,9]
half=len(lst)//2
first_half=lst[:half]
second_half=lst[half:]
print("Sum of first half:", sum(first_half))
print("Sum of second half:", sum(second_half))
```

```python
#Question: WRP in python to delete the element at the index equal to half the lenght of the given list.Print the list before and after deletion.
lst=[2,3,4,5,6,7,8]
print("Before deletion:", lst)
half_index=len(lst)//2
del lst[half_index]
print("After deletion:", lst)
```

```python
# WAP whichh will take a list as a input from a user and create a new list from a given list the new list have all the even numbers from the given list.
lst=input("Enter a list of numbers separated by commas: ").split(",")
even_numbers=[int(num) for num in lst if int(num)%2==0]
print("Even numbers from the given list:", even_numbers)
```

```python
# WAP using list comprehension to create a list of the sums of corresponding elements from two given lists.
list1=[1,2,3,4]
list2=[5,6,7,8]
sums=[a+b for a,b in zip(list1,list2)]
print(sums)
```
