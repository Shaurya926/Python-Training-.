```python
# Question: WRP in python to divide the list into 2 equal halves and print the sum of each half.    
lst=[2,4,6,8,3,5,8,9]
half=len(lst)//2
first_half=lst[:half]
second_half=lst[half:]
print("Sum of first half:", sum(first_half))
print("Sum of second half:", sum(second_half))
```

```python
#Question: WRP in python to delete the element at the index equal to half the lenght of the given list.Print the list before and after deletion.
lst=[2,3,4,5,6,7,8]
print("Before deletion:", lst)
half_index=len(lst)//2
del lst[half_index]
print("After deletion:", lst)
```
