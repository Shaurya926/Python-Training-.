```python
# Arithmetic Operators(+, -, *, /, %, **, //) in Python
# Addition
a = 10
b = 5
sum_result = a + b
print("Addition:", sum_result)  # Output: 15
# Subtraction
sub_result = a - b
print("Subtraction:", sub_result)  # Output: 5
# Multiplication
mul_result = a * b
print("Multiplication:", mul_result)  # Output: 50
# Division
div_result = a / b
print("Division:", div_result)  # Output: 2.0
# Modulus
mod_result = a % b
print("Modulus:", mod_result)  # Output: 0
# Exponentiation
exp_result = a ** b
print("Exponentiation:", exp_result)  # Output: 100000
# Floor Division
floor_div_result = a // b
print("Floor Division:", floor_div_result)  # Output: 2
```

```python
# Assignment Operators in Python
# Simple Assignment (=)
x = 10
print("Simple Assignment (x = 10):", x)  # Output: 10
# Addition Assignment (+=)
y = 5
y += 3
print("Addition Assignment (y += 3):", y)  # Output: 8
# Subtraction Assignment (-=)
z = 15
z -= 5
print("Subtraction Assignment (z -= 5):", z)  # Output: 10
# Multiplication Assignment (*=)
a = 4
a *= 2
print("Multiplication Assignment (a *= 2):", a)  # Output: 8
# Division Assignment (/=)
b = 20
b /= 4
print("Division Assignment (b /= 4):", b)   # Output: 5.0
# Modulus Assignment (%=)
c = 10
c %= 3
print("Modulus Assignment (c %= 3):", c)  # Output: 1
# Exponentiation Assignment (**=)
d = 2
d **= 3
print("Exponentiation Assignment (d **= 3):", d)  # Output: 8
```

```python
# Logical Operators in Python
# AND Operator
x = True
y = False
and_result = x and y    
print("AND Operator (x and y):", and_result)  # Output: False
# OR Operator
or_result = x or y
print("OR Operator (x or y):", or_result)  # Output: True
# NOT Operator
not_result = not x
print("NOT Operator (not x):", not_result)  # Output: False
```

```python
# Relational Comparison Operators (==, !=, >, <, >=, <=) in Python
# Equality
x = 10
y = 5
eq_result = (x == y)
print("Equality (x == y):", eq_result)  # Output: False
# Inequality
neq_result = (x != y)
print("Inequality (x != y):", neq_result)  # Output: True
# Greater Than
gt_result = (x > y)
print("Greater Than (x > y):", gt_result)  # Output: True
# Less Than
lt_result = (x < y)
print("Less Than (x < y):", lt_result)  # Output: False
# Greater Than or Equal To
gte_result = (x >= y)
print("Greater Than or Equal To (x >= y):", gte_result)  # Output: True
# Less Than or Equal To
lte_result = (x <= y)
print("Less Than or Equal To (x <= y):", lte_result)  # Output: False
# Comparing equal values
a = 7
b = 7
equal_values_result = (a == b)
print("Comparing equal values (a == b):", equal_values_result)  # Output
# Comparing different values
different_values_result = (a != b)
print("Comparing different values (a != b):", different_values_result)  # Output: False: True
# Comparing greater than or equal with equal values
gte_equal_result = (a >= b)
print("Comparing greater than or equal with equal values (a >= b):", gte_equal_result)  # Output: True
# Comparing less than with different values
lt_different_result = (a < b)   
print("Comparing less than with different values (a < b):", lt_different_result)  # Output: False
```
