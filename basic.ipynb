```python
n=int(input("Enter a number: "))
for i in range(2,n//2):
    if(n%i==0):
        print("The number is not prime.")
        break
else:
    print("The number is prime.")
```

```python
def reverse_number(num):
    reversed_num = int(str(num)[::-1])
    return reversed_num

n = int(input("Enter a number: "))
print("Reversed number:", reverse_number(n))
```
